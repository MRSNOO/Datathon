```python
# CLEAN THEM WORDS
vendor = pd.read_csv("vendor.csv")

vendor_copy = vendor.copy()
vendor_copy.dropna(inplace=True)
vendor_copy['preprocessed_' + 'MCC_DESCRIPTION'] = vendor_copy['MCC_DESCRIPTION'].str.lower()

tokenizer = RegexpTokenizer('[a-zA-Z]\w+\'?\w*')
vendor_copy['tokenized_' + 'MCC_DESCRIPTION'] = vendor_copy['preprocessed_' + 'MCC_DESCRIPTION'].apply(lambda row: tokenizer.tokenize(row))
  
df_words_clean = vendor_copy
df_words_clean.head()
```

```python
"""
most common words: [('notelsewhereclassified', 60737),
 ('stationery', 42613),
 ('bookstores', 38197),
 ('office', 27157),
 ('directmarketing', 25553),
 ('schoolsupplystores', 22884),
 ('officesupplies', 19729),
 ('printing', 19729),
 ('writingpaper', 16396),
 ('local', 15648)]
 """
```

```python
words_dictionary = Dictionary(df_words_clean.tokenized_MCC_DESCRIPTION)

words_corpus = [words_dictionary.doc2bow(wordlist) for wordlist in df_words_clean.tokenized_MCC_DESCRIPTION]

words_coherence = []
for nb_topics in range(1,36):
    lda = LdaModel(words_corpus, num_topics = nb_topics, id2word = words_dictionary, passes=10)
    cohmodel = CoherenceModel(model=lda, corpus=words_corpus, dictionary=words_dictionary, coherence='u_mass')
    coh = cohmodel.get_coherence()
    words_coherence.append(coh)

# visualize coherence
plt.figure(figsize=(10,5))
plt.plot(range(1,36),words_coherence)
plt.xlabel("Number of Topics")
plt.ylabel("Coherence Score");
```

```python
k = 6
vendor_lda = LdaModel(words_corpus, num_topics = k, id2word = words_dictionary, passes=10)
lda=vendor_lda
nb_topics = k
nb_words = 10

top_words = [[word for word,_ in lda.show_topic(topic_id, topn=50)] for topic_id in range(lda.num_topics)]
top_betas = [[beta for _,beta in lda.show_topic(topic_id, topn=50)] for topic_id in range(lda.num_topics)]

gs  = gridspec.GridSpec(round(math.sqrt(k))+1,round(math.sqrt(k))+1)
gs.update(wspace=1, hspace=0.5)
plt.figure(figsize=(20,10))
for i in range(nb_topics):
    ax = plt.subplot(gs[i])
    plt.barh(range(nb_words), top_betas[i][:nb_words], align='center')
    ax.invert_yaxis()
    ax.set_yticks(range(nb_words))
    ax.set_yticklabels(top_words[i][:nb_words])
    plt.title("Group "+str(i + 1))
```
